```python
class Student:
    def __init__(self, masv, hoten, dtb):
        self.masv = masv
        self.hoten = hoten
        self.dtb = dtb
        
    def __str__(self):
        return f"Masv: {self.masv}, Ho ten: {self.hoten}, diem tb: {self.dtb}"
```

```python
class Queue:
    def __init__(self):
        self.ds = []
        
    def is_empty(self):
        return len(self.ds) == 0
        
    def push(self, data):            
        self.ds.append(data)
        
    def pop(self):
        return self.ds.pop()
```

```python
class Node:
    def __init__(self, data = None):
        self.right = None
        self.left = None
        self.data = data
        
    def add_node(self, items):
        if self.data is None:
            self.data = items
            return True
        
        if self.data.masv == items.masv:
            return False
        
        if items.masv < self.data.masv:
            if self.left is None:
                self.left = Node(items)
                return True
            else:
                return self.left.add_node(items)
        elif items.masv > self.data.masv:
            if self.right is None:
                self.right = Node(items)
                return True
            else:
                return self.right.add_node(items)  
                
    def NLR(self):
        print(self.data)
        
        if self.left:
            self.left.NLR()
            
        if self.right:
            self.right.NLR()     
            
    # 3.Đếm số nút có 2 con của cây 
    def count_2Child(self):
        count = 0
        if self.left and self.right:
            count = 1
           
        if self.left:
            count += self.left.count_2Child()
            
        if self.right:
            count += self.right.count_2Child()
            
        return count
    
    # 4.Tìm các sinh viên có điểm trung bình > 5 lưu vào Queue
    def dtb_bigger_5(self):
        que = Queue()
        
        if self.left:
            que.ds += self.left.dtb_bigger_5()
            
        if self.right:
            que.ds += self.right.dtb_bigger_5()
            
        if self.data and self.data.dtb > 5:
            que.push(self.data)  
            
        return que.ds
    
    # 6.Xoá một node có giá trị là MaSV do người dùng nhập 
    def delete_node(self, masv):
        if masv < self.data.masv and self.left:
            self.left = self.left.delete_node(masv)
        elif masv > self.data.masv and self.right:
            self.right = self.right.delete_node(masv)
        else:
            if self.left is None and self.right is None:
                return None
            
            if self.left is None:
                return self.right
            if self.right is None:
                return self.left
            
            temp = self.right
            while self.left:
                temp = self.left
                
            self.data = temp.data
            self.right = self.right.delete_node(temp.data.masv)
            
        return self
    
    # 7.Tìm những nút có 1 con, in ra file MSSV_FindLeaf.txt 
    def find_node_1_child(self, ds = None):
        if ds is None: ds = []
        
        if self.data:
            if (self.left and not self.right) or (self.right and not self.left):
                ds.append(self.data)
                
            if self.left:
                self.left.find_node_1_child(ds)
                
            if self.right:
                self.right.find_node_1_child(ds)
                
        return ds
```

```python
node = Node()

while True:
    print("==== Menu ====")
    print("1. Đọc file MSSV.txt đổ vô cây nhị phân tìm kiếm")
    print("2. Duyệt và xuất dữ liệu của cây nhị phân tìm kiếm theo thứ tự NLR ")
    print("3. Đếm số nút có 2 con của cây")
    print("4. Tìm các sinh viên có điểm trung bình > 5 lưu vào Queue, in thông tin trong Queue")
    print("5. Xoá một node có giá trị là MaSV do người dùng nhập")
    print("6. Tìm những nút có 1 con, in ra file 22637851_FindLeaf.txt")
    print("7. Thoát")
    
    choice = int(input("Nhập vào lựa chọn của bạn"))
    
    if choice == 1:
        with open("22637851.txt", 'r') as f:
            for i in f:
                # content = f.read()
                items = i.strip().split(',')

                msv = int(items[0])
                hoten = items[1]
                dtb = float(items[2])
                
                student = Student(msv, hoten, dtb)
                
                add_node = node.add_node(student)
                
                if add_node == True:
                    print("Đã thêm thành công")
                else:
                    print("Thêm không thành công")
    elif choice == 2:
        print("Xuất dữ liệu theo NLR")
        node.NLR()
    elif choice == 3:
        node_2_Child = node.count_2Child()
        print(f"Số nút có 2 con của cây: {node_2_Child}")
    elif choice == 4:
        ds_bigger_5 = node.dtb_bigger_5()
        
        print("Các sinh viên có dtb > 5")
        
        for i in ds_bigger_5:
            print(i)
    elif choice == 5:
        masv_delete = int(input("Nhập mssv bạn cần xóa: "))
        dele = node.delete_node(masv_delete)
        
        if dele:
            print("Da xoa thanh cong")
        else:
            print("Xoa khong thanh cong")
    elif choice == 6:
        node_1_child = node.find_node_1_child()
        
        with open("22637851_FindLeaf.txt", 'w') as f:
            for i in node_1_child:
                f.write(str(i) + '\n')
    else:
        print("Thoát")
        break
```
