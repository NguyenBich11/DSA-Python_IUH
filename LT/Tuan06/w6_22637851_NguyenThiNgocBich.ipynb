```python
class Stack:
    def __init__(self):
        self.stack = []
        
    def is_empty(self):
        return len(self.stack) == 0
    
    def push(self, num):
        self.stack.insert(0, num)
        
    def pop(self):
        return self.stack.pop(0)
```

```python
class Queue:
    def __init__(self):
        self.queue = []
        
    def is_empty(self):
        return len(self.queue) == 0
    
    def push(self, num):
        self.queue.append(num)
        
    def pop(self):
        return self.queue.pop()
```

nhạc) vào hệ thống theo lựa chọn của người dùng cho mỗi lần thêm Xóa media (1đ)
- Hiển thị danh sách media (0.5)
- Tìm kiếm media theo ID (Linear Search) 0.5đ
- Cập nhật thông tin media (0.5đ)
- Tìm media mới nhất theo năm phát hành (Binary Search) (1đ)
- Sắp xếp media theo ID sử dụng (Selection Sort) (1đ)
- Lưu trữ dữ liệu media vào file (1đ)
- Đọc dữ liệu media từ file (1đ)

```python
import json

class QuanLy:
    def __init__(self):
        self.ds = Stack()
        
    def addElements(self, element):
        self.ds.push(element)
        
    def removeElement(self):
        if self.ds.is_empty():
            return "Not thing"
        return self.ds.pop()
            
    def show(self):
        if self.ds.is_empty():
            return "Not thing"
        
        for i in self.ds:
            return i
        
    def search_id(self, id):
        for i in self.ds:
            if id == i.id:
                return i
            
        return False
    
    # Cập nhật thông tin media
    def updateMedia(self, id, title, year, category, available):
        result = self.search_id(id)
        if result:
            if title:
                result.title = title
            if year:
                result.year = year
            if category:
                result.category = category
            if available:
                result.available = available
                
            return True #cập nhật thành công
        
        return False
    
    def search_year(self, input_arr, year):
        rs = sorted(input_arr)
        
        head = 0
        tail = len(rs) - 1
        middle = (head + tail) // 2
        
        while head <= tail:
            if rs[middle] == year:
                return 1
            else:
                if rs[middle] > year:
                    tail = middle - 1
                head = middle - 1 
                
                middle = (head + tail) // 2
                
        return -1
             
    # Sắp xếp media theo ID sử dụng (Selection Sort)
    def selectionSort_id(self):
        n = len(self.ds)
    
        for i in range(n):
            min_index = i
            for j in range(i + 1, n):
                if self.ds[j] < self.ds[min_index]:
                    min_index = j
            
            # Đổi chỗ giá trị tại 2 vị trí index
            self.ds[i], self.ds[min_index] = self.ds[min_index], self.ds[i]
            
    def save_to_file(self, filename):
        f_data = []
        temp_stack = Stack()
        while not self.ds.is_empty():
            element = self.ds.pop()
            f_data.append(element.__dict__)
            temp_stack.push(element)
        self.ds = temp_stack
        with open(filename, 'w') as f:
            json.dump(f_data, f, indent=4)
        print("Elemtnt data saved to file.")

    def read_from_file(self, filename):
        try:
            with open(filename, 'r') as f:
                f_data = json.load(f)
                for data in f_data:
                    self.ds.push(data)
            print("Element data read from file.")
        except FileNotFoundError:
            print("File not found.")
            
```
