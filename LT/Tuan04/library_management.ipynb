```python
class Book:
    book_counter = 1
    
    def __init__(self, title, author, publication_year, price, quantity):
        self.id = f"B{Book.book_counter:03d}"
        self.title = title
        self.author = author
        self.publication_year = publication_year
        self.price = price
        self.quantity = quantity
        Book.book_counter += 1
```

```python
class Library():
    def __init__(self):
        self.library = {}
        
    def addBook(self, n):
        for i in range(n):
            title = input("Enter book title: ")
            author = input("Enter author of book: ")
            publication_year = int(input("Enter publication of year: "))
            price = int(input("Enter price of book: "))
            quantity = int(input("Enter quantity of book: "))
            
            new_book = Book(title, author, publication_year, price, quantity)
            self.library[new_book.id] = new_book
            
    def showLibrary(self):
        if not self.library:
            print("Not have book in library")
        for book in self.library.values():
            print(f"Book ID: {book.id}, Title: {book.title}, Author: {book.author}, Publication Year: {book.publication_year}, Price: {book.price}, Quantity: {book.quantity}")
            # book.printBook().globals()

    # Sắp xếp theo: Tên sách (Bubble Sort)
    def sortTitle(self):
        book_list = list(self.library.values())
        n = len(book_list)
        
        for i in range(n):
            for j in range(0, n - i - 1):
                if book_list[j].title > book_list[j + 1].title:
                    book_list[j], book_list[j + 1] = book_list[j + 1], book_list[j]
                    
        self.library = {book.id: book for book in book_list}
                    
        self.showLibrary()
        
    # Sắp xếp theo: Năm xuất bản (Insertion Sort)
    def sortingPublicationYear(self):
        book_list = list(self.library.values())
        n = len(book_list)
        
        for i in range(n):
            key = book_list[i]
            j = i - 1
            
            while j >= 0 and key.publication_year < book_list[j].publication_year:
                book_list[j+1] = book_list[j]
                j -= 1
                
            book_list[j + 1] = key
            
        self.library = {book.id: book for book in book_list}
        
        self.showLibrary()
    
    # Sắp xếp theo: Giá sách (Selection Sort) 
    def sortingPrice(self):
        book_list = list(self.library.values())
        n = len(book_list)
        
        for i in range(n):
            min = i
            for j in range(i + 1, n):
                if book_list[j].price < book_list[min].price:
                    min = j
                    
            book_list[i], book_list[min] = book_list[min], book_list[i]
            
        self.library = {book.id: book for book in book_list}
        self.showLibrary()
             
    # Tìm kiếm sách (theo ID hoặc tên)
    def searchingID_Name(self, keyword):
        keyword = keyword.lower()
        foundBook = []
        
        for book in self.library.values():
            if book.id.lower() == keyword or keyword in book.title.lower():
                foundBook.append(book)
                
        if not foundBook:
            print("Không có sách phù hợp")
        else:
            for book in foundBook:
                self.showLibrary()
                
    # Mượn sách với kiểm tra số lượng
    def borrow_book(self, id):
        if id in self.library:
            book = self.library[id]
            
            if book.quantity > 0:
                book.quantity -= 1
                print(f"✅ Bạn đã mượn sách '{book.title}'. Số lượng còn lại: {book.quantity}")
            else:
                print(f"❌ Sách '{book.title}' đã hết, không thể mượn!")
        else:
            print("Không tìm thấy sách")
            
    # Trả sách với kiểm tra số lượng
    def return_book(self, id):
        if id in self.library:
            book = self.library[id]
            
            book.quantity += 1
            print(f"✅ Bạn đã trả sách '{book.title}'. Số lượng hiện tại: {book.quantity}")
            
        else:
            print("Không tìm thấy sách")
```

```python
library = Library()
```

```python
def menu():
    while True:
        print("Mời bạn lựa chọn:")
        print("1. Thêm sinh viên")
        print("2. Sắp xếp theo tên sách")
        print("3. Sắp xếp theo năm xuất bản")
        print("4. Sắp xếp theo giá sách")
        print("5. Tìm kiếm sách theo id hoặc tên")
        print("6. Mượn sách với ID")
        print("7. Trả sách")
        print("8. Hiển thị sách trong thư viện")
        
        choice = input("Nhập lựa chọn đi: ")
        if choice == "1":
            n = int(input("Nhập số lượng sách muốn thêm: "))
            library.addBook(n)
        elif choice == "2":
            print("------Sắp xếp sách theo title-----")
            library.sortTitle()
        elif choice == '3':
            print("------Sắp xếp sách theo năm xuất bản-----")
            library.sortingPublicationYear()
        elif choice == "4":
            print("------Sắp xếp sách theo giá-----")
            library.sortingPrice()
        elif choice == "5":
            print("------Tìm kiếm sách theo id hoặc tên-----")
            keyword = input("Nhập vào ID hoặc tên sách mà bạn muốn tìm kiếm")
            library.searchingID_Name(keyword)
        elif choice == "6":
            print("------Mượn sách với ID-----")
            id = input("Nhập vào ID muốn mượn")
            library.borrow_book(id)
        elif choice == "7":
            print("------Trả sách với ID-----")
            id = input("Nhập vào ID sách mà bạn muốn trả")
            library.return_book(id)
        elif choice == "8":
            library.showLibrary()
        else:  
            break
```

```python
menu()
```
