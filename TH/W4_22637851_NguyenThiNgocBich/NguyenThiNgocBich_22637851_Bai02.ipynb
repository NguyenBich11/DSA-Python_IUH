```python
class Student:
    def __init__(self, MaSV, HoTen, SDT, Diem):
        self.MaSV = MaSV
        self.HoTen = HoTen
        self.SDT = SDT
        self.Diem = Diem

    def printStudent(self):
        return f"Mã số sinh viên: {self.MaSV}, Họ tên: {self.HoTen}, Số điện thoại: {self.SDT}, Diem: {self.Diem}"
```

```python
import json

class StudentManagement:
    def __init__(self):
        self.lst = []

    def addStudent(self, n):
        for i in range(n):
            id = int(input("Enter ID of student: "))
            fullname = input("Enter Full name of student: ")
            phoneNumber = input("Enter phone number of student: ")
            mark = float(input("Enter mark of student: "))

            student = Student(id, fullname, phoneNumber, mark)
            self.lst.append(student)

    def showAllStudent(self):
        for student in self.lst:
            if not student:
                print("Not have student")
            else:
                print(student.printStudent())
    
    def searchLinear(self, mssv):
        for i in self.lst:
            if i.MaSV == mssv:
                print("======Result searching======")
                print(i.printStudent())
            else:
                print("Have not student")
                
    def searchBinary(self, phoneNumber):
        head = 0
        tail = len(self.lst) - 1
        
        while head <= tail:
            mid = (head + tail) // 2
            
            if self.lst[mid].SDT == phoneNumber:
                return self.lst[mid]
            else:
                if self.lst[mid].SDT > phoneNumber:
                    tail = mid - 1
                else:
                    head = mid + 1

                mid = (head + tail) // 2
                
        return None
    
    def searchMaxMark(self):
        if not self.lst:
            print("Not have student")
            return
        
        mark = [student.Diem for student in self.lst]
        maxs = max(mark)
        
        markMax = [student for student in self.lst if student.Diem == maxs]
        
        print("======Mark largets======")
        if markMax:
            for student in markMax:
                print(student.printStudent())
        else:
            print("Not have Student")
            
    def searchMaxMark_SecondMark(self):
        if not self.lst:
            print("Not have student")
            return
        
        sort_lst = sorted(self.lst, key = lambda mark: self.lst.Diem, reverse=True)
        
        if len(sort_lst) >= 2:
            return sort_lst[0], sort_lst[1]
        elif len(sort_lst) == 1:
            return sort_lst[0], "Not have student has mark second!"
        else:
            return "List of student is not enough 2 students."
        
    def bubbleSortByMaSV(self):
        n = len(self.lst)
        for i in range(n):
            for j in range(0, n-i-1):
                if self.lst[j].MaSV > self.lst[j+1].MaSV:
                    self.lst[j], self.lst[j+1] = self.lst[j+1], self.lst[j]

    def insertionSortByHoTen(self):
        for i in range(1, len(self.lst)):
            key = self.lst[i]
            j = i - 1
            while j >= 0 and self.lst[j].HoTen > key.HoTen:
                self.lst[j + 1] = self.lst[j]
                j -= 1
            self.lst[j + 1] = key

    def selectionSortByDiem(self):
        n = len(self.lst)
        for i in range(n):
            min_idx = i
            for j in range(i + 1, n):
                if self.lst[j].Diem < self.lst[min_idx].Diem:
                    min_idx = j
            self.lst[i], self.lst[min_idx] = self.lst[min_idx], self.lst[i]

    def saveToFile(self, filename="students.json"):
        with open(filename, "w") as f:
            json.dump([student.__dict__ for student in self.lst], f, indent=4)
        print("Student data saved successfully!")
```

```python
studentSystem = StudentManagement()

while True:
    print("1. Add student")
    print("2. Show all student")
    print("3. Search student by ID(Linear searching)")
    print("4. Search student by phone number(Binary searching)")
    print("5. Search student have max mark")
    print("6. Search student have max mark and second mark")
    print("7. Sort by MaSV (Bubble Sort)")
    print("8. Sort by HoTen (Insertion Sort)")
    print("9. Sort by Diem (Selection Sort)")
    print("10. Save students to file")
    print("0. Exit")

    try:
        choice = input("Enter your choice: ")

        if choice == '1':
            n  = int(input("Enter quality student: "))
            studentSystem.addStudent(n)
        elif choice == '2':
            studentSystem.showAllStudent() 
        elif choice == '3':
            id = int(input("Enter ID of student: "))
            studentSystem.searchLinear(id)
        elif choice == '4':
            phone = input("Enter phone number: ")
            student = studentSystem.searchBinary(phone)
            
            if student:
                print("======Result======")
                print(f"ID of student: {student.MaSV}, Full name: {student.HoTen}, Phone number: {student.SDT}, Mark: {student.Diem}")
            else:
                print(f"Have not student with phone number: {phone}")
        elif choice == '5':
            studentSystem.searchMaxMark()
        elif choice == '6':
            search = studentSystem.searchMaxMark_SecondMark()
            
            if search:
                for student in search:
                    print(f"ID of student: {student.MaSV}, Full name: {student.HoTen}, Phone number: {student.SDT}, Mark: {student.Diem}")
        elif choice == '7':
            studentSystem.bubbleSortByMaSV()
            print("Sorted by MaSV.")
        elif choice == '8':
            studentSystem.insertionSortByHoTen()
            print("Sorted by HoTen.")
        elif choice == '9':
            studentSystem.selectionSortByDiem()
            print("Sorted by Diem.")
        elif choice == '10':
            studentSystem.saveToFile()
        elif choice == '0':
            break
        else:
            print("Choice not right!")
    
    except Exception as e:
        print(f"Error is happen: {e}")
```
