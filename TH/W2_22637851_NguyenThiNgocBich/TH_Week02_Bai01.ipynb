```python
def display_library_menu():
    print("\n=== LIBRARY MANAGEMENT SYSTEM ===")
    print("1. Add new book")
    print("2. Remove book")
    print("3. Borrow book")
    print("4. Return book")
    print("5. Display all books")
    print("6. Search book by ISBN")
    print("7. Exit")
    return input("Enter your choice (1-7): ")
```

```python
class Book:
    def __init__(self, title, author, isbn, status="available"):
        self.title = title  
        self.author = author  
        self.isbn = isbn 
        self.status = status 
```

```python
class Library:
    def __init__(self):
        self.lst_book = []
        
    def add_book(self, book):
        self.lst_book.append(book)
        
    # Xóa sách theo ISBN
    def remove_book(self, isbn):
        for book in self.lst_book:
            if(isbn == book.isbn):
                self.lst_book.remove(book)
                return True
        return False

    # Mượn sách
    def borrow_book(self, isbn):
        for book in self.lst_book:
            if isbn == book.isbn and book.status == 'available':
                book.status = 'borrowed'
                return True
        return False
    
    # Trả sách
    def return_book(self, isbn):
        for book in self.lst_book:
            if isbn == book.isbn and book.status == 'borrowed':
                book.status = 'available'
                return True
        return False
    
    # Hiển thị danh sách sách
    def display_books(self):
        if not self.lst_book:
            print("Not book in library")
        else:
            print("=== List of book ===")
            for book in self.lst_book:
                print("Title: ", book.title, ", Author: ", book.author, ", ISBN: ", book.isbn, ", Status: ", book.status)
            
    # Tìm kiếm theo ISBN
    def search_isbn(self, isbn):
        for book in self.lst_book:
            if(isbn == book.isbn):
                print("Title: ", book.title, ", Author: ", book.author, ", ISBN: ", book.isbn, ", Status: ", book.status)
                return
        print("There is no book with ISBN")
```

```python
if __name__ == "__main__":
    library = Library()
    
    while True:
        choice = display_library_menu()
        
        if choice == '1':
            title = input("Enter title: ")
            author = input("Enter author: ") 
            isbn = input("Enter isbn: ")
            book = Book(title, author, isbn)
            library.add_book(book)
            print("Book added Successfully")
        elif choice == '2':
            isbn = input("Enter ISBN")
            if library.remove_book(isbn):
                print("Book removed Successfully")
            else:
                print("Book not found")
        elif choice == '3':
            isbn = input("Enter ISBN")
            if library.borrow_book(isbn):
                print("Borrow Book Successful")
            else:
                print("Borrow Book Failed")
        elif choice == '4':
            isbn = input("Enter ISBN")
            if library.return_book(isbn):
                print("Return Book Successful")
            else:
                print("Return Book Failed")
        elif choice == '5':
            library.display_books()
        elif choice == '6':
            isbn = input("Enter ISBN")
            library.search_isbn(isbn)
        elif choice == '7':
            print("7. Exit the system")
            break
        else:
            print("Choice not reasonable. Please try again")
```
