```python
class Student:
    def __init__(self, Masv, Hoten, Malop, DiemTB):
        self.Masv = Masv
        self.Hoten = Hoten
        self.Malop = Malop
        self.DiemTB = DiemTB
        
    def __str__(self):
        return f"Masv: {self.Masv}, HoTen: {self.Hoten}, MaLop: {self.Malop}, DiemTB: {self.DiemTB}"
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
    # nhap du lieu cho cay
    def add_node(self, student):
        if self.data.Masv == student.Masv:
            return False
            
        if student.Masv < self.data.Masv:
            if self.left is None:
                self.left = Node(student)
                return True
            else:
                self.left.add_node(student)
        elif student.Masv > self.data.Masv:
            if self.right is None:
                self.right = Node(student)
                return True
            else:
                self.right.add_node(student)       
     
    # duyet va in theo lnr
    def LNR(self):
        if self.left:
            self.left.LNR()
            
        print(self.data)
        
        if self.right:
            self.right.LNR()
    
    # dem so nut la cua cay
    def countLeaf(self):
        if self.left is None and self.right is None:
            return 1
        
        dem = 0
        if self.left:
            dem += self.left.countLeaf()
            
        if self.right:
            dem += self.right.countLeaf()
            
        return dem
    
    # tinh chieu cao cua cay
    def height_tree(self):
        height_left = 0
        height_right = 0
        
        if self.left:
            height_left = self.left.height_tree()
            
        if self.right:
            height_right = self.right.height_tree()
            
        return 1 + max(height_right, height_left)
    
    # timf kiems
    def search_masv(self, masv):
        if self.data.Masv == masv:
            return self
        
        if masv < self.data.Masv and self.left:
            return self.left.search_masv(masv)
        
        if masv > self.data.Masv and self.right:
            return self.right.search_masv(masv)

    # xoas
    def delete_node(self, masv):
        if self.data is None: return None
        
        if masv < self.data:
            self.left.delete_node(masv)
        elif masv > self.data:
            self.right.delete_node(masv)   
        else:
            # nuts khong cos con
            if self.left is None and self.right is None:
                return None
            
            # nust cos 1 con
            if self.left is None:
                return self.right
            if self.right is None:
                return self.left
            
            # nust cos 2 con
            temp = self.right
            while self.left:
                temp = self.left
                
            self.data = temp.data
            self.right = self.right.delete_node(temp.data.Masv)
            
        return self
```
